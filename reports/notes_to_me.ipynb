##Purpose

The purpose of this project is to develop a content-based movie recommendation system that prioritizes relevance based on movie characteristics rather than popularity alone. The goal is to generate recommendations that are similar in content while also taking into account movie quality and the reliability of user ratings.

Unlike commercial systems that often optimize for engagement or exposure, this model aims to provide more balanced and transparent recommendations, where similarity is the primary driver and additional factors are used to refine the results.

Method

The recommendation system is based on a content-based approach using movie genres and user-generated tags. These textual features are combined and transformed into numerical representations using TF-IDF vectorization. Cosine similarity is then used to identify movies with similar content.

To improve the ranking of recommendations, a reranking step is applied. In this step, additional factors are incorporated:

Similarity score (primary signal)
Weighted rating, reflecting overall movie quality
Rating count, used as an indicator of reliability and popularity

Since these variables are on different scales, the weighted rating is normalized to ensure a fair combination. A final score is calculated as a weighted sum of similarity, rating, and popularity, where similarity is given the highest importance.

This approach allows the model to balance content relevance with quality and reliability, while avoiding over-reliance on popularity.

**Why not KNN?*
I considered using K‑Nearest Neighbors (KNN) because it is simple, easy to understand, and works well on small datasets. However, I decided not to use KNN for the full MovieLens dataset. The full dataset is very large, and KNN does not scale well. It requires computing similarities between all users or items, which becomes too slow and uses too much memory on a personal computer. Because of this, KNN is not practical for the full dataset, even though it has clear advantages on smaller ones.

**Why TD-IDF?**
I chose to use a TF‑IDF based content‑based approach instead of K‑Nearest Neighbors. While KNN is simple and effective on small datasets, it does not scale well to the full MovieLens dataset, which contains over 86,000 movies. KNN requires computing pairwise similarities between all items or users, which becomes too slow and memory‑intensive on a personal computer. TF‑IDF, on the other hand, is much more efficient and allows me to compute movie similarities without the heavy computational cost.
